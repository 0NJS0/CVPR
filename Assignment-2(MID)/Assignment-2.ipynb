Importing Library and Dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Loading Dataset

```python
train_data = pd.read_csv(r"C:\Users\NJS\Desktop\archive\mnist_train.csv")
test_data  = pd.read_csv(r"C:\Users\NJS\Desktop\archive\mnist_test.csv")

train_data.head(100)
```

Masking labels above 4

```python
train_data = train_data[train_data.iloc[:, 0] < 4]
test_data = test_data[test_data.iloc[:, 0] < 4]

train_data.head(100)
```

Normalizing data

```python
X_train = train_data.iloc[:, 1:].values / 255.0
X_test = test_data.iloc[:, 1:].values / 255.0

pd.set_option('display.max_columns', None)
display(pd.DataFrame(X_train))
```

One Hot Encoder

```python
def one_hot_encode(labels, num_classes):
    one_hot = np.zeros((labels.size, num_classes))
    one_hot[np.arange(labels.size), labels] = 1
    return one_hot

y_train_labels = train_data.iloc[:, 0].values
Y_train = one_hot_encode(y_train_labels, num_classes=4)

y_test_labels = test_data.iloc[:, 0].values
Y_test = one_hot_encode(y_test_labels, num_classes=4)
```

Neural Network

```python
class NeuralNetwork:
    def __init__(self):
        self.inputLayer_Neurons = 784
        self.hiddenLayer1_Neurons = 24
        self.hiddenLayer2_Neurons = 12
        self.hiddenLayer3_Neurons = 6
        self.outputLayer_Neurons = 4

        self.learning_rate = 0.01

        self.W_H1 = np.random.randn(self.inputLayer_Neurons, self.hiddenLayer1_Neurons)
        self.W_H2 = np.random.randn(self.hiddenLayer1_Neurons, self.hiddenLayer2_Neurons)
        self.W_H3 = np.random.randn(self.hiddenLayer2_Neurons, self.hiddenLayer3_Neurons)
        self.W_O  = np.random.randn(self.hiddenLayer3_Neurons, self.outputLayer_Neurons)

    def relu(self, x, der=False):
      if der:
          return (x > 0).astype(float)
      return np.maximum(0, x)

    def softmax(self, x):
        exp_x = np.exp(x - np.max(x, axis=1, keepdims=True))
        return exp_x / np.sum(exp_x, axis=1, keepdims=True)

    def feedForward(self, X):
        self.X = X
        self.h1 = self.relu(np.dot(X, self.W_H1))
        self.h2 = self.relu(np.dot(self.h1, self.W_H2))
        self.h3 = self.relu(np.dot(self.h2, self.W_H3))
        self.output_input = np.dot(self.h3, self.W_O)
        self.out = self.softmax(self.output_input)
        return self.out

    def backPropagation(self, Y):
            m = Y.shape[0]
            d_out = (self.out - Y) / m

            dW_O = np.dot(self.h3.T, d_out)
            dH3 = np.dot(d_out, self.W_O.T) * self.relu(self.h3, der=True)

            dW_H3 = np.dot(self.h2.T, dH3)
            dH2 = np.dot(dH3, self.W_H3.T) * self.relu(self.h2, der=True)

            dW_H2 = np.dot(self.h1.T, dH2)
            dH1 = np.dot(dH2, self.W_H2.T) * self.relu(self.h1, der=True)

            dW_H1 = np.dot(self.X.T, dH1)


            self.W_O  -= self.learning_rate * dW_O
            self.W_H3 -= self.learning_rate * dW_H3
            self.W_H2 -= self.learning_rate * dW_H2
            self.W_H1 -= self.learning_rate * dW_H1

    def train(self, X, Y):
            self.feedForward(X)
            self.backPropagation(Y)

    def predict(self, X):
            probs = self.feedForward(X)
            return np.argmax(probs, axis=1)
```

Trainning

```python
NN = NeuralNetwork()

losses = []

for epoch in range(10000):
    NN.train(X_train, Y_train)
    if epoch % 100 == 0:
        probs = NN.feedForward(X_train)
        loss = -np.mean(np.sum(Y_train * np.log(probs + 1e-8), axis=1))
        losses.append(loss)
        print(f"Epoch {epoch}: Loss = {loss:.4f}")

plt.plot(losses)
```

```python
preds = NN.predict(X_test)
true = np.argmax(Y_test, axis=1)
acc = np.mean(preds == true)
print(f"\nTest Accuracy: {acc * 100:.4f}%")
```
